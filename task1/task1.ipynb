```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from PIL import Image
img=Image.open("lena.jpg")
display(img)
```

```python
#采用降采样的方法来降低图像的空间分辨率
def down_sample(img):
    h,w=img.shape[:2]
    dst=np.zeros((h//2,w//2))
    dst=img[::2,::2]
    return dst
```

```python
img=cv2.imread("lena.jpg")
#cv2读取的图片默认为BGR，这里转换为RGB
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_1=down_sample(img)
img_2=down_sample(img_1)
img_3=down_sample(img_2)
plt.subplot(221),plt.imshow(img),plt.title("origin"),plt.axis("off")
plt.subplot(222),plt.imshow(img_1),plt.title("downsample_once"),plt.axis("off")
plt.subplot(223),plt.imshow(img_2),plt.title("downsample_twice"),plt.axis("off")
plt.subplot(224),plt.imshow(img_3),plt.title("downsample_thrice"),plt.axis("off")
plt.show()
```

```python
#通过改变灰度值来改变空间分辨率
for i in range(8):
    plt.subplot(2,4,i+1)
    if i<4:
        img_temp=np.uint(img*(2**i))
        plt.imshow(img_temp),plt.axis("off")
    else:
        img_temp=np.uint(img.astype(np.float16)/(2**(8-i)))
        plt.imshow(img_temp),plt.axis("off")
plt.show()
```
