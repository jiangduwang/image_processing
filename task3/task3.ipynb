```python
from PIL import Image
img=Image.open(r'D:\project\python\code\image_operation\lena.jpg')
display(img)
```

```python
from torchvision import transforms
#平移
img1=transforms.RandomAffine(degrees=0,translate=(0.2,0.3))(img)
#旋转
img2=transforms.RandomAffine(degrees=(30,40))(img)
#镜像
#水平翻转
img3=transforms.RandomHorizontalFlip(p=1)(img)
#垂直翻转
img4=transforms.RandomVerticalFlip(p=1)(img)
```

```python
import matplotlib.pyplot as mlt
axis_py=mlt.figure().subplots(3,2)
axis_py[0,0].imshow(img);axis_py[0,0].set_title('src');axis_py[0,0].axis('off')
axis_py[0,1].axis('off')
axis_py[1,0].imshow(img1);axis_py[1,0].set_title('translation');axis_py[1,0].axis('off')
axis_py[1,1].imshow(img2);axis_py[1,1].set_title('rotation');axis_py[1,1].axis('off')
axis_py[2,0].imshow(img3);axis_py[2,0].set_title('horizontalflip');axis_py[2,0].axis('off')
axis_py[2,1].imshow(img4);axis_py[2,1].set_title('verticalflip');axis_py[2,1].axis('off')

```
