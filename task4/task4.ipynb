```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("lena.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
h,w=img.shape[:2]
```

```python
#
img_1=np.fft.fft2(img)
img_2=np.fft.fftshift(img_1)
#
img1=np.fft.ifft2(np.abs(img_1)*np.exp(1j*np.angle(img_1)))
img1=np.real(img1)
#0
img2=np.real(np.fft.ifft2(np.abs(img_1)))
#1
img3=np.real(np.fft.ifft2(np.exp(1j*np.angle(img_1))))
plt.subplot(221),plt.imshow(img,"gray"),plt.title("origin"),plt.axis("off")
plt.subplot(222),plt.imshow(img1,"gray"),plt.title("way1"),plt.axis("off")
plt.subplot(223),plt.imshow(img2,"gray"),plt.title("way2"),plt.axis("off")
plt.subplot(224),plt.imshow(img3,"gray"),plt.title("way3"),plt.axis("off")
plt.show()
```
